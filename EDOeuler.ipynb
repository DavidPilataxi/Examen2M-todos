```python
import numpy as np
import matplotlib.pyplot as plt
from src import ODE_euler  # Importando el código de Euler desde src

# Definimos la ecuación diferencial y' = y - (t^2 + 1)
def f(t, y):
    return y - t**2 + 1

# Solución exacta basada en la imagen
def exact_solution(t):
    return (t + 1) ** 2 - 0.5 * np.exp(t)

# Parámetros del problema
a, b = 0, 2  # Intervalo de tiempo
y_t0 = 0.5  # Condición inicial
N = 10  # Número de pasos

# Resolver usando el método de Euler
ys, ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

# Calcular la solución exacta en un rango más denso
ts_exact = np.linspace(a, b, 100)
ys_exact = exact_solution(ts_exact)

# Graficar resultados
plt.figure(figsize=(8, 5))
plt.plot(ts, ys, 'bo-', label="Euler's Method", markerfacecolor='none')
plt.plot(ts_exact, ys_exact, 'r-', label="Exact Solution")
plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.title("Comparación de Euler con la solución Exacta")
plt.grid()
plt.show()
```

4) Qué pasa cuando N aumenta?  
- Mayor precisión (mejor aproximación a la solución exacta).
-  Menos error de truncamiento (disminuye con ℎ).
- Más puntos en la gráfica (curva más suave).
❌ Mayor costo computacional (más cálculos necesarios).

Conclusión: Más N mejora la precisión pero aumenta el tiempo de cómputo.
